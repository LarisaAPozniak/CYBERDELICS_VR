```python
subjects = ['AS32', 'CB55', 'DE21', 'DF23', 'EE89', 'FK11', 'FK24', 'fx48',
       'GF87', 'GG88', 'GS21', 'HG11', 'JK67', 'JU77', 'KI09', 'KL72',
       'KO33', 'LL90', 'MN99', 'MO45', 'NO21', 'OO11', 'OP01', 'PO67',
       'PP00', 'RD56', 'RO88', 'RT12', 'TR90', 'TY65', 'YT50']
colors = ['PinkBlue', 'Blue', 'Purple', 'GreenPurple', 'Green', 'Red',
       'PurpleGreen', 'Yellow', 'BluePink', 'GreenOrange', 'OrangePink',
       'RedBlue', 'BlueRed', 'PinkYellow']
stims = ['CubesControl', 'Fractal', 'HoneyComb', 'kaleidoscope']
```

```python
list_epochs_cc = []
for subject in subjects:
    for stim in ['CubesControl']:
        for color in colors:
            try:
                fname = f"D:/CYBERDELICS/CLEANED EPOCHS/{subject}_{stim}_{color}_epo.fif"
                epoch = mne.read_epochs(fname, preload=True)
                epoch = epoch.interpolate_bads(reset_bads=True, mode='accurate').crop(tmin = -1, 
                                                                                      tmax = 20).compute_psd(method = 'welch', 
                                                                                                             fmin = 0.1, 
                                                                                                             fmax = 30)
                list_epochs_cc.append(epoch.data)
            except Exception as e:
                print(e)
list_epochs_hc = []
for subject in subjects:
    for stim in ['HoneyComb']:
        for color in colors:
            try:
                fname = f"D:/CYBERDELICS/CLEANED EPOCHS/{subject}_{stim}_{color}_epo.fif"
                epoch = mne.read_epochs(fname, preload=True)
                epoch = epoch.interpolate_bads(reset_bads=True, mode='accurate').crop(tmin = -1, 
                                                                                      tmax = 20).compute_psd(method = 'welch', 
                                                                                                             fmin = 0.1, 
                                                                                                             fmax = 30)
                list_epochs_hc.append(epoch.data)
            except Exception as e:
                print(e)
list_epochs_f = []
for subject in subjects:
    for stim in ['Fractal']:
        for color in colors:
            try:
                fname = f"D:/CYBERDELICS/CLEANED EPOCHS/{subject}_{stim}_{color}_epo.fif"
                epoch = mne.read_epochs(fname, preload=True)
                epoch = epoch.interpolate_bads(reset_bads=True, mode='accurate').crop(tmin = -1, 
                                                                                      tmax = 20).compute_psd(method = 'welch', 
                                                                                                             fmin = 0.1, 
                                                                                                             fmax = 30)
                list_epochs_f.append(epoch.data)
            except Exception as e:
                print(e)
list_epochs_k = []
for subject in subjects:
    for stim in ['kaleidoscope']:
        for color in colors:
            try:
                fname = f"D:/CYBERDELICS/CLEANED EPOCHS/{subject}_{stim}_{color}_epo.fif"
                epoch = mne.read_epochs(fname, preload=True)
                epoch = epoch.interpolate_bads(reset_bads=True, mode='accurate').crop(tmin = -1, 
                                                                                      tmax = 20).compute_psd(method = 'welch', 
                                                                                                             fmin = 0.1, 
                                                                                                             fmax = 30)
                list_epochs_k.append(epoch.data)
            except Exception as e:
                print(e)
    
    
            
            
```

```python

x = np.linspace(0.1, 30, 122)  # frequency axis

# Frequency bands (example: delta, theta, alpha, beta, gamma)
bands = {
    'delta': (0.1, 4),
    'theta': (4, 8),
    'alpha': (8, 12),
    'beta': (12, 30),
}
band_colors = {
    'delta': '#e0f7fa',
    'theta': '#e0fce0',
    'alpha': '#fff3e0',
    'beta': '#f3e5f5',
    'gamma': '#fbe9e7',
}

# Set up plot
plt.figure(figsize=(6, 4), dpi=600)
k =np.array(list_epochs_k).mean(axis = 0).mean(axis = 0).mean(axis = 0)
f =np.array(list_epochs_f).mean(axis = 0).mean(axis = 0).mean(axis = 0)
cc =np.array(list_epochs_cc).mean(axis = 0).mean(axis = 0).mean(axis = 0)
hc =np.array(list_epochs_hc).mean(axis = 0).mean(axis = 0).mean(axis = 0)
fig = plt.figure()

spectra = [cc, hc]
labels = [ 'cc', 'hc']
# Plot semi-transparent frequency band fills
for band, (fmin, fmax) in bands.items():
    plt.axvspan(fmin, fmax, color=band_colors[band], alpha=0.2, label=f'{band} band')

# Plot each spectrum with SEM
colrs = ['green', 'magenta']
for i, y in enumerate(spectra):
    print(i)
    plt.plot(x, y, color=colrs[i], lw=1,  label=f'Spectrum {labels[i]}')
    plt.fill_between(x, 0, y , color=colrs[i], alpha=0.4)



# Style the plot
plt.xlabel('Frequency (Hz)', fontsize=12)
plt.ylabel('Power, mV2/Hz', fontsize=12)
plt.title('Spectral Power Density', fontsize=14, weight='bold')
plt.xlim([x.min(), x.max()])
#plt.ylim(bottom=0, top = 0.000000000011)
plt.legend(frameon=False, fontsize=8, loc='upper right', ncol=1)
plt.grid(False)

# Remove top/right spines
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

# Tight layout for publication
plt.tight_layout()
plt.show()
```
