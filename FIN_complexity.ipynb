```python
import numpy as np
from scipy import signal
from scipy.signal import hilbert
from random import shuffle

def Pre(X):
    '''
    Detrend and normalize input data. X is a 2D NumPy array (samples × timepoints)
    '''
    ro, co = X.shape
    Z = np.zeros((ro, co))
    for i in range(ro):
        Z[i, :] = signal.detrend(X[i, :] - np.mean(X[i, :]))
    return Z

def cpr(string):
    '''
    Lempel-Ziv-Welch compression of binary input string.
    Returns the dictionary size.
    '''
    d = {}
    w = ''
    for c in string:
        wc = w + c
        if wc in d:
            w = wc
        else:
            d[wc] = wc
            w = c
    return len(d)

def str_col(X):
    '''
    Convert continuous time series matrix X into a binary string
    by Hilbert transform + thresholding.
    '''
    ro, co = X.shape
    M = np.abs(hilbert(X, axis=1))
    TH = np.mean(M, axis=1)

    s = ''
    for j in range(co):
        for i in range(ro):
            s += '1' if M[i, j] > TH[i] else '0'

    return s

def LZc(X):
    '''
    Compute normalized Lempel-Ziv complexity (LZc) of X
    '''
    Xp = Pre(X)
    SC = str_col(Xp)
    M = list(SC)
    shuffle(M)
    shuffled = ''.join(M)
    return cpr(SC) / float(cpr(shuffled))
```
