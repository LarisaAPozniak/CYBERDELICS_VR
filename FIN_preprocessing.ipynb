```python
import mne
import os
import os.path as op
import numpy as np
import pandas as pd
from mne.preprocessing import ICA, corrmap, create_ecg_epochs, create_eog_epochs
from mne.preprocessing import ICA
from mne.preprocessing import (
    compute_proj_eog,
    create_eog_epochs,
)
from mne.preprocessing import annotate_muscle_zscore

#pip install mne_qt_browser
mne.viz.set_browser_backend('qt')  # Enable mne-qt-browser backend if mne < 1.0
```

```python
subjects = ['AS32', 'CB55', 'DE21', 'DF23', 'EE89', 'FK11', 'FK24', 'fx48',
       'GF87', 'GG88', 'GS21', 'HG11', 'JK67', 'JU77', 'KI09', 'KL72',
       'KO33', 'LL90', 'MN99', 'MO45', 'NO21', 'OO11', 'OP01', 'PO67',
       'PP00', 'RD56', 'RO88', 'RT12', 'TR90', 'TY65', 'YT50']

for subject in subjects:
    raw = mne.io.read_raw_brainvision(f'D:\hse\psychodelic_like_experience\отдельный разговор\{subject}\{subject}.vhdr', preload = True)
    raw.set_channel_types({'VEOG': 'eog', 'HEOG' : 'eog', 'mark': 'stim', 'pletism':'misc', 'KGR':'misc'})
    raw = raw.rename_channels(mapping = {'FP2' : 'Fp2'})
    raw.set_montage(montage='standard_1020')
    raw = raw.filter(l_freq=1, h_freq=30, picks='eeg')
    raw.plot()  # Wait for user input before moving to the next step

    eog_projs, _ = compute_proj_eog(
        raw, n_grad=0, n_mag=0, n_eeg=2, reject=None, no_proj=True
    )
    raw_clean = raw.add_proj(eog_projs).copy()
    raw_clean.plot(block=True)
    # Show the filter weights in a topomap
    #weights.plot()
    #raw_clean.plot()
    ica = mne.preprocessing.ICA(
        n_components=25,  max_iter="auto", random_state=99
    )
    ica.fit(raw_clean)
    ica.plot_components()
    ica.plot_sources(raw_clean, block=True)
    reconst_raw = raw_clean.copy()
    ica.apply(reconst_raw)
    reconst_raw.plot()
    reconst_raw.save(f'D:/hse/psychodelic_like_experience/data_processing/eeg_cleaned_ica/{subject}cleaned_raw.fif', overwrite=True)
```
